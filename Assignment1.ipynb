```python
# Define the bin edges
bins = [0, 45, 55, float('inf')]  # 0-45, 45-55, 55+

# Define the corresponding labels for the bins
labels = ['45-', '45+', '55+']

# Use pd.cut to create the new binned column
dfs['age_group'] = pd.cut(dfs['age'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
dfs.head()
```

```python
# Define the bin edges
bins = [0, 200, 240, float('inf')]  # 0-45, 45-55, 55+

# Define the corresponding labels for the bins
labels = ['Good', 'Borderline', 'High'] #	Total cholesterol	HDL cholesterol
#Good	Less than 200 (the lower, the better)	Ideal is 60 or higher; but, 40+ for males or 50+ for females is acceptable
#Borderline to moderately elevated	200239	n/a
#High	240 or higher	60 or higher
#Low	n/a	less than 40 for men and less than 50 for women

# Use pd.cut to create the new binned column
dfs['cholestrol_group'] = pd.cut(dfs['cholesterol'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
dfs.head()
```

```python
# Define the bin edges
bins = [0, 120, 130, float('inf')]  # 0-45, 45-55, 55+

# Define the corresponding labels for the bins
labels = ['Normal', 'Elevated', 'HBP'] #Category	Systolic	Diastolic
#Normal	Less than 120	Less than 80
#Elevated	120129	Less than 80
#High Blood Pressure (Hypertension) Stage 1	130139	80  89
#High Blood Pressure (Hypertension) Stage 2	140 or higher	90 or higher(https://www.baptisthealth.com/blog/heart-care/healthy-blood-pressure-by-age-and-gender-chart)

# Use pd.cut to create the new binned column
dfs['HBP_groups'] = pd.cut(dfs['resting_blood_pressure'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
dfs.head()
```

```python
# Define the bin edges
bins = [0, 170, float('inf')]  # 0-45, 45-55, 55+

# Define the corresponding labels for the bins
labels = ['Normal', 'High']

# Use pd.cut to create the new binned column
dfs['heart_rate'] = pd.cut(dfs['max_heart_rate_achieved'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
dfs.head()
```

```python
# Define the bin edges for the grouping
bins = [0.0, 1.0, 2.0, float('inf')]  # 0.0 - 1.0 (mild), 1.0 - 2.0 (moderate), 2.0+ (severe)

# Define the corresponding labels
labels = ['Mild', 'Moderate', 'Severe']

# Use pd.cut to create the new grouped column
dfs['st_depression_group'] = pd.cut(dfs['st_depression'], bins=bins, labels=labels, right=False)

# Display the updated DataFrame
dfs.head()
```
